```python
#Probability
class Probability:
    def __init__(self,a:list)->None:
        self.a=a
    def get_probability(self,favourable_event: int | float, total_events:int | float)->float:
        return float(favourable_event)/float(total_events)

p1= Probability([0,1,2,3])
print(p1.get_probability(3,8))
```

```python
#Permutations
def factorial(n:int)->int:
    ret=1
    for i in range(n):
        ret*= (i+1)
        #print(ret)
    return ret
print(factorial(5))
def permut(n:int,r:int)->float:
    return float(factorial(n)/factorial(n-r))
def combinat(n:int,r:int)->float:
    return float(permut(n,r)/factorial(r))
print(combinat(2,0))
```
